```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
mat = loadmat("ex6data3.mat")
mat = loadmat("ex6data3.mat")
X = mat["X"]
y = mat["y"]
Xval = mat["Xval"]
yval = mat["yval"]
```

```python
m,n = X.shape[0],X.shape[1]
pos,neg= (y==1).reshape(m,1), (y==0).reshape(m,1)
plt.scatter(X[pos[:,0],0],X[pos[:,0],1],c="r",marker="+",s=25)
plt.scatter(X[neg[:,0],0],X[neg[:,0],1],c="y",marker="o",s=25)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel="rbf", C = 10, gamma = 10)
classifier.fit(X,y.flatten())
```

```python
classifier.score(X, y.flatten())
```

```python
classifier.score(Xval, yval.flatten())
```
